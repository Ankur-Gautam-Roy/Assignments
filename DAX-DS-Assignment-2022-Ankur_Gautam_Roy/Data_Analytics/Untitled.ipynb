```python
import matplotlib.pyplot as plt
import csv
```

```python
x = []
y = []
  
with open('healthcare-dataset-stroke-data.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    for row in lines:
        x.append(row[1])
        y.append(row[9])
  
plt.plot(x, y, color = 'g', linestyle = 'dashed',
         marker = 'o',label = "HealthCare")
  
plt.xticks(rotation = 20)
plt.xlabel('ID')
plt.ylabel('Average glucose level')
plt.title('Gender-Age report', fontsize = 15)
plt.grid()
plt.legend()
plt.show()
```

```python
total_bmi = len(df['bmi']);
check_nan = df['bmi'].isnull().sum();
if( check_nan > 0):
    percentage_bmi = ((check_nan)/total_bmi)*100;
    print("Has Null values and are: ", check_nan);
    print("Percentage of Null values are: ",percentage_bmi,"%"); 
else:
    print("No Null values");

```
